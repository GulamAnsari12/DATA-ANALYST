```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
#countplot : Survival base on Gender
sns.countplot(df, x = 'survived', hue= 'sex', palette = 'RdBu')
plt.title('Survived Vs not Survived by Gender')
plt.show()
```

```python
sns.barplot(df, x= 'class' , y='fare', hue= 'sex')
plt.title('Average Fare by Passenger Class and Gender')
plt.show()
```

```python
sns.violinplot(df, x = 'class' , y = 'sex', hue = 'class')
plt.title('Class distributon by Gender')
plt.show()
```

```python
sns.pointplot(df, x = 'class', y = 'survived', hue = 'sex', markers = ['o', 'x'])
plt.title('Survival Rate By Class and Gender')
plt.show()
```
