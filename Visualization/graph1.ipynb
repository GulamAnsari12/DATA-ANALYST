```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
#countplot : Survival base on Gender
sns.countplot(df, x = 'survived', hue= 'sex', palette = '')
plt.title('Survived Vs not Survived by Gender')
plt.show()
```
