```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
ag = sns.load_dataset("penguins")
ag
```

```python
#Histogram Chart
sns.displot(ag["bill_length_mm"],bins = [35,40,45,50,55,60,65], kde = True, color = 'red')
plt.show()
```

```python
#Barplot
sns.barplot(x = ag.island , y = ag.bill_length_mm ,)
plt.show()
```

```python
#Scatter Plot
sns.scatterplot(x = "bill_length_mm",
                y = "bill_depth_mm" , data = ag , hue = "sex", style = "sex", size = "sex", sizes = (100,40))
plt.show()
```

```python

#HeatMap
ag1 = np.linspace(1,10,20).reshape(4,5)   #2d array banakar heatmap use kiye h
ag1
```

```python
ag2 = sns.load_dataset("tips")
ag2
```

```python
#Count Plot
sns.countplot(x = "sex", data = ag2)
plt.show()
```

```python
#Pair Plot
sns.pairplot(ag2, hue = "sex")
plt.show()
```

```python
#Box Plot
sns.set(style = "whitegrid")
sns.boxplot(x = "day", y = "total_bill", data = ag2, hue = "sex", palette='dark:red')
plt.show()
```

```python
#Factor Plot
sns.catplot(x = "tip", y = "size", data = ag2, hue ="sex")
plt.show()
```

```python
#Violin Plot
sns.violinplot(x = "day", y = "total_bill", data = ag2 , hue = "day", palette = "dark:red", linewidth = 3)
plt.show()
```

```python
#Strip Plot
sns.stripplot(x = "day", y = "total_bill", data = ag2, hue = "day",)
plt.show()
```
